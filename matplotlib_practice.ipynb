```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of the strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with the plt.show() function
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("Pyber Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create lables for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("Pyber Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Create the plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth=2, label="Boston")
# Create and set x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Create a title
ax.set_title("Pyber Fare by Month")
# Add a y axis limit
ax.set_ylim(0, 45)
# Add a grid
ax.grid()
```

```python
# Add explode values and increase size of pie chart
plt.subplots(figsize=(8,8))
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode, labels=x_axis, autopct="%.1f%%")
plt.show()
```
